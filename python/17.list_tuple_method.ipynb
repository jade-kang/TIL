# 리스트와 튜플 응용하기 - 메소드

### 리스트에 요소 추가하기 - append / extend / insert

- append: 요소 하나를 추가
- extend: 리스트를 연결하여 확장
- insert: '특정 인덱스'에 요소 추가 (append와 extend는 리스트 끝에 추가한다.)

```python
# 요소 추가하기 
a = [] 
a.append(1)  # 요소 하나를 추가

b = [2, 3]
b.extend([4, 5]) 

# append로 리스트에 리스트 추가하기 
c = []
c.append(b)  # 리스트를 연결하여 확장 

print(a)  # [1]
print(b)  # [2, 3, 4, 5]
print(c)  # [[2, 3, 4, 5]]
```

```python
# insert로 특정 인덱스에 요소 추가하기
d = [1, 2, 3]
d.insert(1, 4)  # 인덱스 1에 4 추가
print(d)  # [1, 4, 2, 3]

# insert로 리스트에 리스트 추가하기
d.insert(2, [5, 6])  # 인덱스 2에 [5, 6] 추가
print(d)  # [1, 4,
```

- insert(0, 요소) # 리스트의 맨 처음에 요소 넣기
- insert(len(리스트), 요소) #리스트의 맨끝에 요소 넣기 

### 리스트에서 요소 제거하기 - pop / remove

- pop: 마지막 요소 또는 특정 인덱스의 요소를 삭제 (del 과 동일하다)
- remove: 특정 값을 찾아서 삭제

```python
a = [10, 20, 30]
a.pop(1) #삭제한 값을 반환한다
```

```python
a = [10, 20, 30]
a.remove(20)
a
```

```python
# 삭제하려는 요소가 여러개 있으면 맨 앞의 것을 없앤다
a = [10, 20, 30, 20]
a.remove(20)
a
```

### 리스트로 스택과 큐 만들기

- 스택 ?


- 위의 컵 사진처럼, a, b, c가 스택을 이루면, 꺼낼땐 c, b, a 순서대로 나와야 한다. 
- 이는 곧, append로 스택을 채우고, pop으로 순차적으로 꺼낸다 
    - LIFO (last in, first out, 즉 나중에 들어온 놈이 먼저 나가는 구조)
- 컨트롤 z 처럼, 나중에 들어온 데이터를 먼저 처리하는 형식에서 사용

- 큐 ? 


- queue(큐)? = 줄서기 
    - 먼저 들어온 사람이 먼저 나간다 (FIFO)
    - 먼저 들어온 데이터를 먼저 처리할 때사용 

### 리스트에서 특정 값의 인덱스 구하기 ? - index

```python
a = [10, 20, 30, 20]
a.index(20)  # 20의 인덱스 찾기
```

### 특정 값의 개수 구하기 - count

```python
a = [10, 20, 30, 20]
a.count(20)
```

### 리스트의 순서 뒤집기 - reverse

```python
a = [10, 20, 30, 20]
a.reverse()  # 리스트의 순서 뒤집기
a
```

### 리스트의 순서 정렬하기 - sort()

- sort() 또는 sort(reverse=False): 리스트의 값을 작은 순서대로 정렬(오름차순)
- sort(reverse=True): 리스트의 값을 큰 순서대로 정렬(내림차순)

```python
a = [10, 20, 30, 20]
a.sort(reverse=False)  # 오름차순 정렬
a
```

### sort vs sorted ?

- sort는 메소드를 사용한 리스트를 정렬
- sorted는 정렬된 '새 리스트'를 정의한다. 

```python
a = [10, 20, 30, 20]
a.sort()  # 리스트의 값을 작은 순서대로 정렬(오름차순)
a
```

```python
a = [10, 20, 30, 20]
b = sorted(a)
b 
```

### 리스트의 모든 요소 제거 - clear()

```python
a = [10, 20, 30, 20]
a.clear()  # 리스트의 모든 요소 제거
a
```

```python
a = [10, 20, 30, 20]
del a[:]
a 
```
