### 숫자

- 정수(integer) : 자연수(1,2,3,,)와 자연수의 음수(-1,-2,,), 0 을 모두 포함하는 수의 집합
- 소수(floating point) : 부동소수점 붙은 숫자들 
- 복소수는 공학에서 사용 


### 연산자

- / : 나눗셈 (실수로 결과 나옴)
- // : 나눗셈 후 몫만 반환 (소수점 이하 버림)
- % : 나머지만 반환 
- ** : 거듭제곱
- *divmod(,)* : 몫과 나머지 함께 반환  
- 값을 정수로 만드는 법 
    - int(숫자)
    - int(계산식)
    - int('문자열') 
        int('10') = 10

```python
print(4/3)  
```

```python
print(4/2)  #나머지가 없어도 실수로 반환
```

```python
print(divmod(5,2)) #튜플 형식으로 몫, 나머지 함께 반환
```

```python
quotient, remainder = divmod(5,2) 
print(quotient, remainder)
```

```python
# 실수와 정수를 계산하면 실수가 나온다 . 표현 범위가 넓은 쪽으로 계산되기 때문이다
print(4.0+5)
```

### 2진수, 8진수, 16진수의 표현

- 정수는 10진수 이외에도 2진수, 8진수, 16진수로도 표현할 수 있다. 
- 2진수 : 숫자 앞에 0b를 붙이며, 0과 1 사용 
- 8진수 : 숫자 앞에 0o를 붙이며, 0부터 7까지 사용
- 16진수 : 숫자 앞에 0x를 붙이며, 0부터 9, A부터 F까지 사용

```python
0b110
```

```python
0b1001  #2^0 + 2^3 = 9
```

```python
0o10 
```

```python
0x11
```
