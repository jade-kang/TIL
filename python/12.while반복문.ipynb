### for문과 while문의 차이점

- while 반복문은 반복 횟수가 정해져 있지 않을 때 자주 사용 
- for 문은 반복횟수가 정해져있을 때 자주 사용 

### while 반복문으로 100번 출력하기 

- while 반복문은 **조건식**으로만 동작한다
- 반복할 코드 내에 조건식에 영향을 주는 **변화식**이 들어간다

- 초기식부터 시작하여 조건식을 판별 
- 이때, 조건식이 참이면 반복할 코드와 변화식을 함께 수행 
- 그리고 다시 조건식을 판별하여 참이면 코드 반복, 거짓이면 반복문을 끝내고 다음 코드를 실행한다

```python
i = 0                     # 초기식
while i < 100:            # while 조건식  #반복하다 이 조건식에 부합하지 않으면 다음 코드로 넘어간다.
     print('Hello, world!')    # 반복할 코드 
     i += 1                    # 변화식
```

```python
#초깃값을 1부터 시작 

i = 1 #초기식 
while i < 100: #조건식
    print('hello', i) #반복할 코드 
    i += 1 #변화식 
```

```python
#1부터 시작해서 홀수만 출력하도록 ? 
i = 1
while i < 100: 
    print('hello', i)
    i += 2 #변화식 (2씩 증가)
```

- 초깃값 감소시키기 

```python
i = 100 #초기식
while i > 0: 
    print('hello', i)
    i -= 1
```

- 입력한 횟수대로 반복하기

```python
#count = int(input())
count = 10
i = 0
while i < count:
    print('hello', i)
    i += 1  
```

### 반복 횟수가 정해지지 않은 경우 ?

- 난수를 생성해서 숫자에 따라 반복을 끝내 보자 
- 난수(random number)란 특정 주기로 반복되지 않으며, ***규칙 없이 무작위로 나열되는 숫자***를 뜻한다. 
- ex) 주사위를 굴려 나온 숫자

---

- python에서 난수를 생성하려면 random 모듈 필요 
- import 키워드를 사용하여 가져올 수 있다. 

```python
import random #random 모듈을 가져온다
```

```python
#random.random() 으로 random 모듈의 random 함수 호출해오기 
random.random()  #실행할 때마다 계속 다른 실수가 출력되는 데, 이를 난수라고 한다
```

- randint 함수는 난수를 생성할 범위를 지정하며, 범위에 지정한 숫자도 난수 
    - random.randint(a,b)

```python
#randint 함수로 주사위 만들기 
random.randint(1,6)
```

- 1과 6사이 난수를 생성한 뒤 3이 나오면 반복을 끝내는 코드 

```python
import random 
 
i = 0 #초기식 #i 라는 값은 계속 새로운 난수로 되어야 하기 때문에 while문 안에 넣는다. 
while i != 3: #3이 아닐 때 계속 반복 
    i = random.randint(1,6) #randint를 사용하여 1과 6 사이 난수 생성
    print(i) 
```

```python
import random 
#이렇게 초기식(i=0)이 없으면 while 루프문의 시작조차 x 
while i != 3: 
    i = random.randint(1,6)
    print(i)
```

### 무한 루프

```python
while True: #조건식이 만족되면 계속 실행
    print('hello') #문자열,, 계속 참이여서 계속 출력됨. 
```

- while에 조건식 대신 true를 지정하면 무한 루프 만들어짐
- 조건식이 항상 참(true) 이므로 변화식도 필요없음


### 연습문제 

- 변수 두 개를 다르게 반복하기
- 다음 소스 코드를 완성하여 정수 2 5, 4 4, 8 3, 16 2, 32 1이 각 줄에 출력되게 만드세요. while에 조건식은 두 개 지정하고, 두 변수를 모두 변화시켜야 합니다.

```python
i = 2 
j = 5 

while i <= 32 or j >= 1:
    print(i, j)
    i *= 2
    j -= 1
```

### 연습문제 : 교통카드 잔액 계산하기 

- 표준 입력으로 금액(정수)이 입력됩니다. 1회당 요금은 1,350원이고, 교통카드를 사용했을 때마다의 잔액을 각 줄에 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). 단, 최초 금액은 출력하지 않아야 합니다. 그리고 잔액은 음수가 될 수 없으며 잔액이 부족하면 출력을 끝냅니다.

- 입력
- 10000
- 결과
- 8650
- 7300
- 5950
- 4600
- 3250
- 1900
- 550

```python
#내 오답 

#money = int(input())
money = 10000 

while money >= 0:
    money -= 1350 #교통카드 사용 
    print(money) #잔액 출력 
```

```python
#정답 : 교통카드를 사용할 수 있는 최소 잔액 이상 돈이 있을 때까지 loop가 돌아가야함
money = 10000

while money >= 1350:
    money -= 1350
    print(money)
```
