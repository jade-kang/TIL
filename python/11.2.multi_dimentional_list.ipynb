# 반복문에서 다차원 배열의 처리 

- 1차원 리스트만으로는 많은 정보들을 담을 수 없다 
    - ex) ['junkyu', 'min', 'woojin'] 

- 여러 정보들을 담기 위해서는 리스트 내에 리스트를 추가해야 한다. 
    - 2차원 리스트 : [['jun', 'seoul', 'man'], ['min', 'busan', 'man'], ['woo', 'incheon', 'woman']]
    - 이런식으로 차원을 늘려가면 아무리 많은 데이터들이라도 리스트에 모두 담을 수 있다. 

- 이런 다차원 리스트는 반복문에서 어떻게 처리하는가 ?

```python
persons = [
    ['jun', 'seoul', 'man'], 
    ['min', 'busan', 'man'], 
    ['woo', 'incheon', 'woman']
    ]

print(persons[0][0])
```

```python
# 반복문으로 위의 정보들을 모두 꺼내보자 

for person in persons:  #이렇게 하면, 'person' 은 리스트 내의 각각의 리스트들을 가리킨다. 
    print(person[0]) #person[0] 은 각각의 리스트의 첫번째 요소를 가리킨다. ex) 'jun', 'min', 'woo'
```

```python
for person in persons:
    print(person[0] + ',' + person[1] + ',' + person[2]) #이렇게 하면, 각각의 리스트의 모든 요소들을 꺼낼 수 있다.
```

```python
#위의 과정 쉽게 만들기 (할당하기)

name, city, sex = ['jun', 'seoul', 'man']

#이 방식을 응용하면 숫자 인덱스로 할당하는 것보다 훨씬 간편하게 다차원 리스트를 처리할 수 있다. 
```

- 아래의 코드 더 쉽게 만들기 

- for person in persons:
    - print(person[0] + ',' + person[1] + ',' + person[2]) 

```python
# 인덱스가 아닌, 변수 이름으로 직관적으로 코드 작성하기 

for name, city, sex in persons:
    print(name + ',' + city + ',' + sex) #이렇게 하면, 각각의 리스트의 모든 요소들을 꺼낼 수 있다. 
```

- 응용 문제 


```python
students = [
    ["Alice", 85, 90, 88],
    ["Bob", 78, 81, 85],
    ["Charlie", 92, 88, 95]
]

# 반복문으로 평균 점수 만들기 -> 학생 리스트만 추가로 더 생기면 코드 수정할거없이 수행 가능해짐 

for name, score1, score2, score3 in students:
    average = (score1 + score2 + score3) / 3
    print(name + " 의 평균 점수는" + str(average) + '입니다')
    #소수점 처리하는 방법 
    print(f'{name}의 평규점수는 {average:.2f}입니다')  #소수점 둘째자리까지 출력
```

- 응용문제2 
    - 다음 데이터에서 사람 이름과 나이만 출력 
    - 출력 예시
        - tom - 25 
        - jane - 30 


```python
people = [
    ("Tom", 25, "Seoul"),
    ("Jane", 30, "Busan"),
    ("Mike", 27, "Incheon")
]

for name, age, city in people:
    print(name + "-" + str(age))
```
