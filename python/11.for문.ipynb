### for와 range 사용하기

- 반복문은 반복 횟수, 반복 및 정지 조건을 자유 자재로 설정 가능하다. 


```python
# for 변수 in range(횟수):
#     반복할 코드 

#
```

- for i in range(100)을 하면, 내부적으로 0부터 99까지로 i값이 설정됨. 
    - i는 loop index 라고 불리며, index의 i를 사용 . (다른거 써도 문제없다)

- 순서도 
    - 이런 형태를 'loop'라고 부른다

```python
# for 반복문으로 hello world 10번 출력하기
for i in range(10): #range안에 반복할 횟수를 설정
    print("hello world") #반복할 코드 
```

```python
# i를 0~9까지 잡고, 반복하며, 각각의 i도 반환하기 
for i in range(10):
    print(i, "hello world") #반복할 코드
```

### for와 range 응용하기 

- 시작하는 숫자와 끝나는 숫자 지정하기 
    - range에 횟수만 지정하면 숫자가 0부터 시작되지만, 다음과 같이 시작, 끝나는 숫자를 지정해 반복할 수도 있음. 

```python
for i in range(5, 15): #5부터 15까지 반복
    print(i, "hello world") #반복할 코드
```

- 증가폭 사용하기 
    - range는 증가폭 설정 가능하다. 0부터 9까지 숫자 중 '짝수'만 출력해보자

```python
for i in range(0,10,2):
    print(i, "hello world") #반복할 코드
```

- 숫자를 감소시키기
    1. 증가폭을 음수로 지정 
    2. reversed를 사용하면 숫자의 순서를 뒤집을 수 있다

```python
for i in range(10,0,-1): #10부터 1까지 반복
    print(i, "hello world") #반복할 코드
```

```python
#9부터 0까지 홀수들만 출력하는 법 ?
for i in range(9, -1, -2):
    print(i, "hello world") #반복할 코드
```

```python
#reversed를 사용하여 숫자를 감소시키기
for i in reversed(range(10)):  #그냥 range(10)으로 하면 0부터 9까지 출력되지만, reversed를 사용하면 9부터 0까지 출력된다
    print(i, "hello world") #반복할 코드
```

```python
for i in reversed(range(1,10)): #원래는 1부터 9까지지만, reversed를 사용하면 9부터 1까지 출력된다   
    print(i, "hello world")
```

- 입력한 횟수대로 반복하기

```python
count = int(input("반복할 횟수를 입력하세요: "))

for i in range(count): #사용자가 입력한 횟수만큼 반복
    print(i, "hello world") 
```

### 시퀀스 객체로 반복하기 

- for에 range 대신 시퀀스 객체들을 사용해도 문제없음 
    - for는 리스트 , 튜플 , 문자열 등 시퀀스 객체로 반복할 수 있음
    - for에 range대신 리스트를 넣으면 리스트의 요소를 꺼내면서 반복함 .

```python
a = [1, 2, 3, 4, 5]
for i in a: #리스트 a의 요소를 하나씩 꺼내서 i에 할당
    print(i, "hello world") #반복할 코드
```

- 튜플도 마찬가지로 튜플의 요소를 꺼내며 반복

```python
fruits = ("apple", "banana", "cherry")
for fruit in fruits: #튜플 fruits의 요소를 하나씩 꺼내서 fruit에 할당
    print(fruit, "is a fruit") #반복할 코드
```

- 문자열도 시퀀스 객체임. for에 문자열을 지정하면 문자를 하나씩 꺼내며 반복 가능

```python
for letter in 'python':
    print(letter, end='  ')
```

- 문자열을 뒤집어서 출력하기
    - reversed 사용 -> 시퀀스 객체를 뒤집어 줌 

```python
for letter in reversed('python'):
    print(letter, end = ' ')
```

### 연습문제 : 리스트의 요소에 10을 곱해서 출력하기

```python
x = [49, -17, 25, 102, 8, 62, 21]
for n in x:
    print(n*10, end=' ')  # 리스트 x의 각 요소에 10을 곱해서 출력
```

### 연습문제 : 구구단 출력

- 표준 입력으로 정수가 입력됩니다. 입력된 정수의 구구단을 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). 출력 형식은 숫자 * 숫자 = 숫자처럼 만들고 숫자와 *, = 사이는 공백을 한 칸 띄웁니다.

```python
#x = int(input())
x = 3
for i in range(1,10):
    print(x, '*', i , '=', x*i)
```
