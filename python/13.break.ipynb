### while 에서 break로 반복문 끝내기

- while 무한 루프에서 숫자를 증가시키다 변수 i가 20일 때 반복문을 끝내도록 만들어보자

```python
i = 0 
while True: #무한 루프 
    print(i)
    i += 1 
    if i == 20: #i가 20 일때  
        break  #반복문 끝냄. while의 제어 흐름 벗어나게 됨 

print('끝!')
```

### for 에서 break로 반복문 끝내기 

```python
for i in range(10000): #0부터 9999까지 반복
    print(i)
    if i == 20: #i가 20일때
        break 
print('끝')
```
