### 인덱스 사용하기 

- indexing 
    - 시퀀스 객체의 각 요소는 순서가 정해져 있으며, 이 순서를 '인덱스'라고 한다. 
    - 시퀀스객체[인덱스] - []안에 각 요소의 인덱스를 지정하여 해당 요소에 접근 가능
    - 

```python
a = [30,10,20,50]
a[0] 
```

```python
a[-1] #가장 마지막 요소에 접근
```

```python
a[-3] 
```

```python
hello = "world"
hello[3]
```

- 인덱스의 범위를 벗어나면 ?
    - indexerror 발생 

- 마지막 요소에 접근하는 법 
    - -1 로 인덱스 지정
    - len 함수로 길이 구한 뒤, 이 길이를 인덱스로 지정할 때 참고 
        - len이 4 이면 마지막은 3으로 인덱싱 

```python
a = [1,2,3,4,56]
print(len(a))
print(a[4])
```

### 요소에 새로운 값 할당하기 


- ***시퀀스 객체는 []로 요소에 접근하여 = 로 값을 할당한다***
    - a[0] = 3 
        - 0번째로 인덱싱하여 3을 할당한다 

```python
a = [0,2,3,4,5]
a[0] = 38
a[1] = 32
a[2] = 55
a
```

- ***시퀀스 자료형 중 튜플, range, 문자열은 읽기 전용이다 !!*** 
    - 인덱싱을 통해 새로운 값을 할당할 수 없다. 

- 튜플에 새로운 값 할당시키면 에러난다 

```python
b = (0,0,0,0)
b[0] = 2
```

- range와 문자열도 저장된 요소를 변경할 수 없다. 

```python
r = range(0,10,2)
r[0] = 3 
```

```python
hello = 'python'
hello[0]= 'i'
```

### 시퀀스 객체의 요소 삭제하기 - del 

```python
a = [30,21,23,5,6]
del a[2]
a #2번째 위치의 23이 삭제된다.
```

```python
b = (0,1,2,3)
del b[2] #읽기 전용이다. 
```
