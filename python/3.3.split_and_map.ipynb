### 입력값을 변수 두 개에 저장하기

- input 한 번에 값을 여러 개 입력 받으려면 ?
    - input에 split을 사용하여 변수 여러 개에 저장한다
    - 이때, split은 각 변수를 콤마로 구분한다. 

- 변수1, 변수2 = input().split()
- 변수1, 변수2 = input().split('기준 문자열') #구분 문자열. 입력없으면 콤마로 구분


### 두 숫자의 합 구하기 

- input으로 받은 입력값은 숫자든 문자든 str 형태로 값을 받는다
- 문제점 : 정수 변환 없이 값을 서로 더하면 문자열처럼 이어 붙은 형태로 값이 반환된다 


- int를 통해 type을 변환해준다



### map 을 사용하여 정수로 변환하기 

- map 함수 ?
    - 여러 요소들에 대해 특정한 함수를 한번에 적용시킬 때 사용 

- map에 int와 input().split()을 넣으면 split의 결과를 모두 int 로 변환해준다. 
    - map을 사용하지 않으면 각각의 변수에 대해 a = int(a), b=int(b) 처럼 각각 변환.. 귀찮다

    - 변수1 ,변수2 = map(int, input().split())
    - 실수로 변환하고 싶으면 map(float, input().split())

### 입력받은 값을 콤마 기준으로 분리하기 

- split에 기준 문자열로 콤마를 지정하여, 공백이 아닌 다른 문자로 분리
