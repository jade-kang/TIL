### 연습 문제: 온라인 할인 쿠폰 시스템 만들기


- 표준 입력으로 가격(정수)과 쿠폰 이름이 각 줄에 입력됩니다. Cash3000 쿠폰은 3,000원, Cash5000 쿠폰은 5,000원을 할인합니다. 쿠폰에 따라 할인된 가격을 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다).

```python
price = int(input())
coupon = input()

if coupon == "Cash3000":
    price -= 3000
if coupon == "Cash5000":
    price -= 5000
print(price) 
```

### else , elif

- if 조건문은 분기(branch) 를 위한, 즉 프로그램을 둘 이상으로 나누는 것을 의미 

- if에 else를 사용하면 조건식이 만족할 때와 만족하지 않을 때 각각 다른 코드 실행할 수 있음 
- else는 if 조건문 뒤에 오며 단독 사용 x 


### if 조건문의 동작방식

- True는 if의 코드가 실행, false는 else의 코드가 실행된다
- none은 false로 취급되므로 else의 코드가 실행된다. 

```python
if True:
    print("참")
else:
    print("거짓")

if False:
    print("참")
else:
    print("거짓")

if None:
    print("참")
else:
    print("거짓")
```

- 숫자는 0만 거짓, 나머지는 참임 (정수(2진수,10진수 등등), 실수)

```python
if 0:
    print("참")
else:
    print("거짓")

if 0.1:
    print("참") 
else:
    print("거짓")
```

- 문자열은 내용이 있으면 참, 빈문자열만 거짓 

```python
if 'hello':
    print("참") 
else:
    print("거짓")

if '':
    print("참")
else:
    print("거짓")
```

- 불값 True와 true로 취급하는 10진수, 16진수, 2진수, 실수, 문자열로 if 동작시킴
- 값 자체가 있으면 if는 동작함

- 반대로 *** 0, none, '' 은 FALSE 취급하므로 ELSE가 동작한다 ***

```python
if not '':   #'' = false  / not false = true 
    print(True)   
else:
    print(False)
```

### 연습문제 

- 합격 여부 판단하기
    - A 기업의 입사 시험은 필기 시험 점수가 80점 이상이면서 코딩 시험을 통과해야 합격이라고 정했습니다(코딩 시험 통과 여부는 True, False로 구분). 다음 소스 코드를 완성하여 '합격', '불합격'이 출력되게 만드세요.

```python
written_test = 75 
coding_test = True 

if written_test >= 80 and coding_test == True:
    print("합격")
else:
    print("불합격")
```

- 표준 입력으로 국어, 영어, 수학, 과학 점수가 입력됩니다. 여기서 네 과목의 평균 점수가 80점 이상일 때 합격이라고 정했습니다. 평균 점수에 따라 '합격', '불합격'을 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). 단, 점수는 0점부터 100점까지만 입력받을 수 있으며 범위를 벗어났다면 '잘못된 점수'를 출력하고 합격, 불합격 여부는 출력하지 않아야 합니다.
- 입력
    - 89 72 93 82
- 결과
    - 합격

```python
#point : 점수가 0~100 범위를 벗어날 경우 '잘못된 점수' 라고 출력 

korean, english, math, science = map(int, input().split())
#평균 점수를 구해야 함 
avg = (korean + english + math + science) / 4
#일차적으로 0~100 사이의 범위에 들어갈 경우, 평균을 계산할 수 있게 한다 
if 0<=korean<=100 and 0<=english<=100 and 0<=math<=100 and 0<=science<=100:
    if avg >= 80: 
        print("합격")
    else:
        print("불합격")
#0~100 사에의 범위에 들어가지 않을 경우 '잘못된 점수'로 산출 
else: 
    print("잘못된 점수")
```
