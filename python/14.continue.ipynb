### continue로 코드 실행 건너뛰기

- break와의 차이점 ??
    - break는 반복문 자체에서 탈출을 함 (완전히 끝내버림)
    - continue는 해당하는 일부 코드를 건너뛰지만, 반복문은 계속해서 진행

- for에서 continue로 코드 실행 건너뛰기 
    - for 로 0부터 99까지 반복하면서 홀수만 출력하기

```python
for i in range(100):
    if i % 2 == 0: #i가 짝수일 때
        continue #아래 코드를 '실행하지 않고 건너뜀' 
    print(i) #i가 홀수일 때 
```

- while반복문에서 continue도 동작은 같다

```python
i = 0 #초기식  
while i < 50: #i가 50보다 작을 때 반복 
    i += 1 
    if i % 2 == 0: #i가 짝수일때 
        continue 
    print(i) #i가 홀수일때만 출력
```

### 입력한 횟수대로 반복하기 

```python
#count = int(input("몇 번 반복할까요? ")) #사용자에게 반복 횟수 입력 받기
 
count = 10
i = 0 #초기식 
while True:  #무한 반복 
    print(i) 
    i += 1 
    if i >= count:
        break 

print("반복이 끝났습니다.") #반복이 끝났음을 알림
```

- 입력한 숫자까지 해당하는 홀수를 출력해보자 

```python
#count = int(input("몇 번 반복할까요? ")) #사용자에게 반복 횟수 입력 받기
 
count = 10 
i = 0 
for i in range(count+1):  #0부터 증가해서 count까지 반복(count+1)
    if i % 2 == 0: #i가 짝수일때 , continue 시키기. break x 
        continue  #i가 짝수이면 이 코드 건너 뜀 
    print(i) #i가 홀수일때만 출력 
```

- 1부터 10까지 출력하면서 3의 배수는 제외하는 법 ? 

```python
for i in range(1,11): #1부터 10까지 반복 
    if i % 3 == 0:
        continue #i가 3의 배수이면 코드 건너뛰기 
    print(i)   
```

### 연습문제 : 3으로 끝나는 숫자만 출력하기 

- 0과 73 사이의 숫자 중 3으로 끝나는 숫자만 출력되게 만들기 
    - 3, 13, 23, 33, 43, 53, 63 

```python
#오답 : 아래의 경우, i가 3으로 끝날때만 print(i), i+=1 이 실행되기 때문에 애초에 i가 증가하지 않음 
i = 0  
while True: #무한 반복 
    if i % 10 == 3: #i가 10으로 나누면 3으로 끝나는 숫자 
        print(i)
        i+=1 
    if i > 73: 
        break 
print("반복이 끝났습니다.")
```

```python
i = 0 
while True: 
    if i > 73:
        break 
    if i % 10 == 3: 
        print(i)
    i += 1
print("반복이 끝났습니다.")  # 반복이 끝났음을 알림
```

```python
#오답 
i = 0 
while True: 
    if i > 73:
        break 
    if i % 10 == 3: 
        print(i)
        i += 1 #이게 if 문 내에 존재하면, i가 3으로 끝날때만 i가 증가하게 됨 = 무한반복 
print("반복이 끝났습니다.")
```

```python
#for 문으로 구현 
for i in range(74): 
    if i % 10 == 3: 
        print(i)
```

### 연습문제 : 두 수 사이의 숫자 중 3으로 끝나지 않는 숫자 출력하기 

- 표준 입력으로 정수 두 개가 입력됩니다(첫 번째 입력 값의 범위는 1~200, 두 번째 입력 값의 범위는 10~200이며 첫 번째 입력 값은 두 번째 입력 값보다 항상 작습니다). 다음 소스 코드를 완성하여 첫 번째 정수와 두 번째 정수 사이의 숫자 중 3으로 끝나지 않는 숫자가 출력되게 만드세요. 정답에 코드를 작성할 때는 while True:에 맞춰서 들여쓰기를 해주세요.

```python
# start, stop = map(int, input().split())
start = 8
stop = 45 

i = start 
while True: 
    if i > stop: 
        break 
    if i % 10 != 3: 
        print(i)
    i+=1 
```

```python
#오답 
i = start 
while True:
    if i > stop:
        break 
    if i % 10 == 3: 
        continue  #여기서 continue를 만나면, 아래 코드들이 실행되지 않아, i는 계속 13에 머무른다
    print(i, end=' ')
    i += 1
```

```python
#continue는 아무튼 아래의 코드는 '건너뛴다' 
i = start 
while True: 
    if i > stop: 
        break 
    if i % 10 == 3: 
        i += 1
        continue #i가 3으로 끝나면 건너뛰기 
    print(i, end=' ') #i가 3으로 끝나지 않으면 출력 
    i += 1
```
