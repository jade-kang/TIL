### 알고리즘과 코딩 

- 알고리즘 ?
    - 문제에서 일정한 패턴을 발견하고, 그 패턴을 토대로 문제를 해결하는 절차를 표현하는 것 
- 코딩 ? 
    - 그러한 알고리즘을 컴퓨터의 명령으로 작성하는 행동 (프로그래밍=코딩)
    - 프로그래밍 
        - 컴퓨터 명령어로 표현한다는 뜻 (컴퓨터 쪽에 가까운 표현)
    - 코딩 
        - 파이썬과 같은 프로그래밍 언어로 코드를 작성한다는 의미 (언어 쪽에 가까운 표현)

### REPL

- WHAT? 
    - 파이썬의 출력 방식
    - 코드를 '읽고'(READ), '평가'하고(Evaluation)(계산 실행), '출력'(print)하는 것이 loop 형식으로 
    - Read-Eval-Pring Loop

### python IDLE?

- integrated development and learning environment
- 파이썬을 위한 통합 개발 환경 


- IDLE에서 에러가 났을 경우 
    - 에러난 코드에 커서를 놓고 엔터를 치면 다시 수정할 수 있게 나옴 

- 소스 파일 ?
    - 프로그래밍에서 코드를 파일 형태로 저장한 것을 소스코드 또는 소스파일이라고 함 

- 명령프롬프트에서 출력하기
    - 명령프롬프트 들어가고 "python"이라고 입력을 해서 파이썬을 실행시켜야, 명령 프롬프트 (>>>)가 나옴
