### 시퀀스 자료형

- 리스트, 튜플, range, 문자열의 공통점 ? 
    - ***연속성*** (sequence)
    
    

- 시퀀스 자료형 특징 
    - ***공통 동작과 기능 제공***
- 시퀀스 객체 : 시퀀스 자료형으로 만든 객체 
- 요소(element) : 시퀀스 객체에 들어있는 각 값 

- \+객체 ? 
    - 파이썬의 모든 것은 객체로 이루어짐. 
    - 객체란, 값이나 기능을 담고 있는 상자 정도로 생각 
        - a = 5 일때, 5는 그냥 숫자가 아닌 ,int라는 타입을 가진 객체임. 
            - 이 객체는 숫자 계산, 크기 비교 등의 '기능(메서드)'를 갖고 있음
        - b = [1,2,3] 에서 [1,2,3] 역시 리스트라는 객체이고, append, pop, sort 등의 기능(메서드)를 가지고 있음 

### 시퀀스 내에 특정 값이 있는지 확인하기 
    - ex : 리스트 a에 30이 있는지 확인하기 등등 

```python
a = list(range(0,100,10))
30 in a , 100 in a #튜플 
```

```python
# 없냐 ? = not in 
100 not in a 
```

```python
30 not in a 
```

- 튜플, range, 문자열도 같은 방법으로 특정값 확인 가능

```python
43 in (38, 12, 43)
```

```python
1 in range(10)
```

```python
'p' in 'python'
```

### 시퀀스 객체 연결하기 

- \+ 연산자로 두 객체를 연결해 새 객체 만들기 가능

```python
a = [10,20]
b = [1,2,3,4]
a+b
```

```python
#시퀀스 자료형 중 range는 + 연산자로 객체 연결 안됨
range(10) + range(0,5)
```

```python
#range를 리스트 또는 튜플로 만들어서 연결하면 된다. 
list(range(0,10)) + list(range(10,20,3))
```

```python
#문자열은 +연산자로 여러 문자열을 연결할 수 있음 
'hello' + ' ' + 'world'
```

### 시퀀스 객체 반복하기 

- *연산자 : 시퀀스 객체를 특정 횟수만큼 반복 
    - 0 또는 음수를 곱하면 빈객체가 나온다
    - 실수는 곱하기 불가능 

```python
[0,10,20] * 3
```

```python
[0,10,20] * 1.0 #오류난다 
```

### 시퀀스 객체의 '요소' 개수 구하기 - len

- len() 사용하면 시퀀스 객체의 요소 갯수 나옴
    - 문자열 길이도 곧 요소들의 갯수

```python
a = list(range(0,10))
len(a)
```

```python
b = tuple(range(0,5))
len(b)
```

```python
len(range(0,10,2))  # 0,2,4,6,8 이기 때문에 5 나옴
```

- 문자열의 길이 구하기 (=문자 갯수 구하기)

```python
a = 'hello python'
len(a)
```
