### 중첩루프

- i를 사용하는 바깥쪽 루프는 세로 방향(줄)을 처리하고, j를 사용하는 안쪽 루프는 가로방향을 처리함 

```python
for i in range(5): #세로방향으로 5번 반복 (0~4)
    for j in range(5): #가로 방향으로 
        print('j:', j, sep='', end=' ') #j값 출력 / sep=''는 j:와 j값 사이에 공백을 넣지 않음 / end=' '는 출력 후 다음 줄로 넘어가지 않고, 공백으로 끝남
    print('i:', i, '\\n', sep='') #가로방향으로 숫자를 모두 출력한 후 다음 줄로 넘어감 
```

### 사격형으로 별 출력하기 

- 중첩된 반복문을 사용하여 5x5 사격형 형태로 별을 그려보자

```python
for i in range(5): 
    for j in range(5):
        print('*', end=' ') #줄바꿈 대신 end를 통해 공백 처리 
    print() #각 줄의 끝에서 줄바꿈을 위해 print() 사용
```

```python
# 7x3 사각형 형태로 별 출력 
for i in range(3):
    for j in range(7):
        print('*', end='')
    print()
```

### 계단식으로 별 출력하기

- 계단식으로 별이 하나씩 증가하면서 출력하려면 어떻게 해야할까?? 
    - 사각형으로 별 출력과 달리, 별을 출력하지 않는 부분이 있음. 
    - if조건문으로 제어해야 한다.

```python
# i = 0 일때, j = 0
# i = 1 일때, j = 0, 1 
# i = 2 일때, j = 0, 1, 2
for i in range(5): 
    for j in range(5): 
        if j <= i:  # i값보다 j값이 작거나 같을 때만 별을 출력
            print('*', end=' ')
    print()
```

### 대각선으로 별 출력하기

```python
#오답 : 앞에 공백들을 고려하지 않음 
# i = 0 일때, j = 0 
# i = 1 일때, j = 1 
# i = 2 일때, j = 2 
for i in range(5):
    for j in range(5):
        if j == i:
            print('*', end=' ')
    print() #각 줄의 끝에서 줄바꿈을 위해 print() 사용
```

```python
#정답 : i==j가 아닐땐, 공백으로 채워라 

for i in range(5):
    for j in range(5): 
        if j == i:
            print('*', end=' ')
        else: 
            print(' ', end=' ')
    print()
```

### 역삼각형 모양으로 별 출력하기

```python
# i = 0 일때, j = 0,1,2,3,4
# i = 1 일때, j = 1,2,3,4
# i = 2 일때, j = 2,3,4
# i = 3 일때, j = 3,4
for i in range(5): 
    for j in range(5): 
        if j < i:
            print(' ', end=' ')
        else :
            print('*', end=' ')
    print()
```

### 산모양 별 출력하기 

```python
#    *  
#   ***
#  *****
# *******
#*********
# 가운데를 기준으로 양쪽에 공백 존재 
# 공백 : 위에서부터 4, 3, 2, 1, 0 개 
# 별 : 1, 3, 5, 7, 9 개 (2*줄번호 + 1)  

#오답 : rows - 1 하면 당연히 앞에 공백 4개가 고정된다. . 
rows = 5 
for i in range(1, rows + 1): #i는 1부터 5 
    for j in range(rows - 1): 
        print(' ', end='') #rows - 1 만큼 공백 출력  
    for k in range(2 * i - 1): 
        print('*', end='')
    print()
```

```python
rows = 5 
for i in range(1, rows+1):
    for j in range(rows - i): #i가 증가하면 공백 줄어듬 
        print(' ', end='')
    for k in range(2 * i - 1): #i가 증가하면 별이 늘어남
        print('*', end='')
    print()
```

```python
#reversed를 사용해서 작성해보기
# reversed : 리스트, range 같은 반복 가능한 객체를 거꾸로 뒤집어 순회 
 
rows = 5 
for i in range(rows): #i는 0부터 rows-1 까지 (0,1,2,3,4) 
    for j in reversed(range(rows)): #j 는 rows-1 부터 0까지 (4,3,2,1,0)
        if j > i: 
            print(' ', end='') #i < j 일때는(앞부분) 공백 , 아니면 별 
        else: 
            print('*', end=' ') 
    print() #각 줄의 끝에서 줄바꿈을 위해 print() 사용
```
